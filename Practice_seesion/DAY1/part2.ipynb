```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

#for reproducibility
np.random.seed(42)

#Generating Synthetic customer data
n_samples = 500

age = np.random.randint(18,70,n_samples)
income = np.random.normal(60000,15000,n_samples)
spending_score = np.random.randint(1,100,n_samples)
orders = np.random.poisson(10,n_samples)  

#create dataframe

df =pd.DataFrame({
    "Age": age,
    "Annual_Income" :income,
    "Spending_Score" :spending_score,
    "Orders_count": orders
    })
print(df.head())
```

```python
scaler = StandardScaler()
x_scaled=scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans =KMeans(n_clusters=4,init='random',random_state=42)
kmeans.fit(x_scaled)

labels_kmeans = kmeans.labels_
inertia_kmeans = kmeans.inertia_
sil_kmeans=silhouette_score(x_scaled,labels_kmeans)

print("Kmeans Inertia:",inertia_kmeans)
print("KMeans Silhouette:",sil_kmeans)
```

```python
import numpy as np

def cosine_sim(a,b):
    a=np.array(a)
    b=np.array(b)
    return np.dot(a,b)/(np.linalg.norm(a) * np.linalg.norm(b))
def l2_dis(a,b):
    return np.linalg.norm(np.array(a)-np.array(b))  #np.lingalg.norm

v1=[1,2,3]
v2=[4,5,6]
print("Cosine:",cosine_sim(v1,v2))
print("L2:",l2_dis(v1,v2))
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline

# 1. Sample Documents
documents = [
    "RAG combines retrieval and generation.",
    "Sentence embeddings capture semantic meaning.",
    "Cosine similarity measures vector direction similarity.",
    "TensorFlow and PyTorch are deep learning frameworks."
]

embed_model = SentenceTransformer("all-MiniLM-L6-v2")

doc_embeddings=embed_model.encode(documents)
dim=doc_embeddings.shape[1]
index=faiss.IndexFlatL2(dim)
index.add(np.array(doc_embeddings))

query = "What is RAG in AI?"
query_embedding= embed_model.encode([query])

k=2
distances,indices=index.search(np.array(query_embedding),k)
retrieved_docs = [documents[i] for i in indices[0]]

context ="\n".join(retrieved_docs)
prompt= f"""
Answer the question using the context below:
context:{context}
Question: {query}
"""

generator = pipeline("text-generation",model="gpt2")
response = generator(prompt,max_length=150)

print(response[0]['generated_text'])

```

```python
def chunk_text(text,chunk_size=100):
    words = text.split()
    chunks = []
    for i in range(0,len(words),chunk_size):
        chunk = " ".join(words[i:i+chunk_size])
        chunks.append(chunk)
    return chunks
# print(chunk_text(documents,100))
```
