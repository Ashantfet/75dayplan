```python
import pandas as pd
import json

# -------------------------
# 1️⃣ Create Dummy Data
# -------------------------

data = [
    {"id": 1, "name": "Ashant", "email": "ashant@gmail.com"},
    {"id": 2, "name": "Ravi", "email": "ravi@gmail.com"},
    {"id": 3, "name": "Priya", "email": "priya@gmail.com"},
]

df = pd.DataFrame(data)

# -------------------------
# 2️⃣ Save as CSV
# -------------------------
df.to_csv("data.csv", index=False)

# -------------------------
# 3️⃣ Save as JSON
# -------------------------
df.to_json("data.json", orient="records", indent=4)

# -------------------------
# 4️⃣ Save as Excel
# -------------------------
df.to_excel("data.xlsx", index=False)

print("Dummy files created successfully!")
```

```python
# Read files
df_csv = pd.read_csv("data.csv")
df_json = pd.read_json("data.json")
df_xlsx = pd.read_excel("data.xlsx")

# Merge
df_merged = pd.concat([df_csv, df_json, df_xlsx], ignore_index=True)

print("Merged DataFrame:")
print(df_merged)
```

```python
df_unique = df_merged.drop_duplicates()

print("After Removing Duplicates:")
print(df_unique)
```
