```python
import torch
```

```python
x=torch.tensor([1,2,3,4,5])
y=torch.tensor([6,7,8,9,10])

#addition
z1=x+y
z2=torch.add(x,y)
z3=torch.empty(5)
torch.add(x,y,out=z3)
print(z1,z2,z3)

z=x-y
print(z)
z=torch.true_divide(x,y)
print(z)

t=x*y
print(t)

m=torch.zeros(5)
m.add_(x)
print(m)
m+=x

## exponentiation
z=x.pow(2)
print(z)
z=x**2
print(z)
```

```python
#comparison
z=x>0
print(z)

#matrix multiplication
x1=torch.rand(2,5)
x2=torch.rand(5,3)
x3=torch.mm(x1,x2)
print(x3)
x3=x1.mm(x2)
print(x3)

matrix_exp=torch.rand(5,5)
print(matrix_exp)
print(matrix_exp.matrix_power(3))
```

```python
z=x*y
print(z)
z=torch.dot(x,y)
print(z)
```

```python
batch=32
n=10
m=20
p=30
tensor1=torch.rand(batch,n,m)
tensor2=torch.rand(batch,m,p)
out_bmm=torch.bmm(tensor1,tensor2)
print(out_bmm.shape)
print(out_bmm)
```

```python
x1=torch.rand((5,5))
x2=torch.rand((1,5))
z=x1-x2
print(z)
z=x1**x2
print(z)
```

```python
sum_x=torch.sum(x,dim=0)
print(sum_x)
values,indices=torch.max(x,dim=0)
values1,indices1=torch.min(x,dim=0)
print(values,indices)
print(values1,indices1)
abs_x=torch.abs(x)
print(abs_x)
z=torch.argmax(x,dim=0)
print(z)
z=torch.argmin(x,dim=0)
print(z)
mean_x=torch.mean(x.float(),dim=0)
print(mean_x)
z=torch.eq(x,y)
print(z)
sorted_y,indices=torch.sort(y,dim=0,descending=False)
print(sorted_y,indices)
z=torch.clamp(x,min=0,max=5)
print(z)
```

```python
x=torch.tensor([1,0,1,0,1],dtype=torch.bool)
z=torch.any(x)
print(z)
z=torch.all(x)
print(z)
```
