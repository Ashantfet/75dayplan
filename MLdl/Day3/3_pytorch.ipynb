```python
import torch
```

```python
batch_size=10
features=25
x=torch.rand((batch_size,features))
print(x[0].shape)
print(x[:,0].shape)
print(x[2,0:10].shape)
x[0,0]=100
print(x[0])
x=torch.arange(10)
indices=[2,5,8]
print(x[indices])

x=torch.rand((3,5))
rows=torch.tensor([1,0])
cols=torch.tensor([4,0])
print(x.shape)
print(x[rows].shape)
print(x[:,cols].shape)
print(x[rows,cols].shape)
```

```python
x=torch.arange(10)
print(x[(x<3)|(x>7)])
print(x[x.remainder(2)==0])
print(x[x.remainder(2)==1])
```
