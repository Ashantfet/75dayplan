```python
name = "Ashant"
age = 23
is_student = True
height = 1.75

print(name, age, is_student, height)
print(type(name), type(age), type(is_student), type(height))
```

```python
# List (mutable)
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print("List:", fruits)

# Tuple (immutable)
colors = ("red", "green", "blue")
# colors.append("white") not allopwed 
print("Tuple:", colors)

# Set (unique, unordered)
nums = {1, 2, 3, 3, 4,7,6} #gives sorted
print("Set:", nums)

# Dictionary (key-value pairs)
student = {"name": "Ashant", "age": 23, "branch": "CSE"}
print("Dict:", student)
print("Name:", student["name"])
```

```python
marks = 85

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
else:
    print("Grade: C")
```

```python
for i in range(5):
    print("Loop:", i) #from 0 to n-1
for i in range(1,5):
    print("Loop:", i) #from start to n-1
n = 5
while n > 0:
    print("While:", n)
    n -= 1
```

```python
nums = [1, 2, 3, 4, 5]
squares = [x**2 for x in nums]  #squares of a list
evens = [x for x in nums if x % 2 == 0] # even number in a list
print(squares, evens)
```
