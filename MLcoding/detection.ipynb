```python
from ultralytics import YOLO

# Load pretrained YOLOv8 (VERY IMPORTANT LINE)
model = YOLO("yolov8n.pt")   # n = nano (fast)
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("image.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.title("Input Image")
plt.axis("off")
plt.show()
```

```python
results = model(image)
for r in results:
    boxes = r.boxes.xyxy      # bounding boxes
    scores = r.boxes.conf     # confidence
    labels = r.boxes.cls      # class IDs

    print("Boxes:", boxes)
    print("Scores:", scores)
    print("Labels:", labels)
```

```python
annotated = results[0].plot()

plt.imshow(annotated)
plt.title("Detected Objects")
plt.axis("off")
plt.show()
```
